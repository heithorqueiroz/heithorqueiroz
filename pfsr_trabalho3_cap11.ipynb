```python
import math as m
```

```python
#import numpy as np

alfa = list(range(0,360))
alfa
```

```python
# Função para plotagem

def palfa(lista):
  lista=[]
  for i in alfa:
    Palfa = 3/4*(1+m.cos(alfa[i])**2)
    lista.append(Palfa)
  return lista

phi = palfa(alfa)
```

```python
# Plot 

import numpy as np
import matplotlib.pyplot as plt


#r = np.arange(0, 2, 0.01)
#theta = 2 * np.pi * r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(phi, '.')
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(90)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("Função de fase de espalhamento Rayleigh", va='bottom')
plt.show()
```
